```python
coffee_tea_pastry = {
    'Espresso': 1.50, 
    'Americano': 2.00,
    'Caffe latte': 2.25, 
    'Mocha': 2.25,
    'Cappucino': 2.50,
    'Macchiato': 2.50,
    'White tea': 1.50,
    'Green tea': 2.00,
    'Oolong tea': 2.25,
    'Black tea': 2.25,
    'Earl Grey': 2.50,
    'Matcha': 2.50,
    'Croissant': 5.50,
    'Cinnamon roll': 4.50,
    'Banana bread': 4.25,
    'Pumpkin bread': 3.75,
    'Sticky bun': 4.25,
    'Apple turnover': 3.75}

```

```python
def display_big_menu():
    print('\nCoffee')
    for i, key in enumerate(coffee_tea_pastry):
        if i ==6:
            print('\nTea')
        elif i==12:
            print('\nPastries')
        print(f"{key:<25} {f'${coffee_tea_pastry[key]}':>10}")
        

def display_user_option(option):
    ans=''
    if option==1:
        while ans != 'y' and ans !='n':
            ans=input("Would you like to see the menu? Y/N ").lower()
    elif option==2:
        while ans != 'y':
            ans=input("Would you like to order now? Y/N ").lower()
    
    if ans =='y':
        return True
    else:
        return False
    
def ask_for_order():
    item=input("What would you like? ")
    if check_item(item):
        quantity=int(input("How many? "))
        prepare_receipt(item, quantity)

    
def check_item(item):
    if item in coffee_tea_pastry:
        # print(f"{item} found and price is {coffee_tea_pastry[item]}")
        return True
    else:
        print(f"{item} not found")
        return False

    
def prepare_receipt(item, quantity):
    # print(f"People are buying {quantity} {item}")
    if item in myorder:
        myorder[item] += quantity
    else:
        myorder[item]=quantity
    
def display_receipt():
    subtotal=0
    print('\nReceipt\n')
    print('Item                Quantity      $Price')
    for key in myorder:
        print(f"{key:<20} {myorder[key]:5} {myorder[key]*coffee_tea_pastry[key]:>12}")
        subtotal += myorder[key]*coffee_tea_pastry[key]
    print()
    print(f'                    Subtotal {subtotal:>10.2f}')
    tax = subtotal * 0.06
    print(f'                         Tax {tax:>10.2f}')
    print(f'                       Total {subtotal+tax:10.2f}')
    print('\nThank you for visiting MC Cafe')
```

```python
myorder ={}

print("Welcome to MC Cafe")

if display_user_option(1):
    display_big_menu()

if display_user_option(2):
    ask_for_order()

ans=''
while ans!='y' and ans!='n':
    ans=input("\nWould you like to order another item? Y/N ").lower()

    if ans=='y':
        ask_for_order()
        ans=''
    elif ans=='n':
        display_receipt()
        break

```
